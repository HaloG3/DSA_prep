# HeapSort Definition and Complexity Analysis

## Definition
HeapSort is a comparison-based sorting algorithm that uses a binary heap data structure to sort elements. It works by first building a max-heap (or min-heap) from the input array, then repeatedly extracting the maximum (or minimum) element from the heap and placing it at the end (or beginning) of the array. The heap is adjusted after each extraction to maintain the heap property. This process continues until all elements are sorted.

### Steps:
1. **Build a max-heap** from the input array.
2. **Swap the root** (maximum element) with the last element of the heap.
3. **Reduce the heap size** by 1 and heapify the root to restore the heap property.
4. **Repeat** steps 2–3 until the heap size is 1.

HeapSort is an in-place sorting algorithm but is not stable (i.e., it does not preserve the relative order of equal elements).

## Time Complexity
| Operation         | Time Complexity |
|-------------------|-----------------|
| **Best Case**     | O(n log n)      |
| **Average Case**  | O(n log n)      |
| **Worst Case**    | O(n log n)      |

- **Building the max-heap**: O(n)
- **Heapify operations**: O(log n) per extraction, performed n times, leading to O(n log n).
- **Total**: The overall time complexity is O(n log n) for all cases, as the algorithm's performance is consistent regardless of input distribution.

## Space Complexity
| Operation         | Space Complexity |
|-------------------|------------------|
| **Auxiliary Space** | O(1)            |

- HeapSort is an **in-place algorithm**, requiring no extra space for storing the array beyond a few variables for indexing and swapping.
- However, the recursive nature of heapify may use O(log n) stack space in the worst case for the call stack, but this is typically considered negligible.
- **Total Space Complexity**: O(1) auxiliary space, excluding the input array.

## Notes
- HeapSort is efficient for large datasets due to its consistent O(n log n) time complexity.
- It is not stable, so it may not be suitable for applications requiring preservation of equal elements' order.
- HeapSort is less commonly used in practice compared to QuickSort or MergeSort due to its higher constant factors and lack of stability.

````python
# heap Sort  t: O(1) constant space is possible via swapping, but this is complex 
def heapsort(arr): # its a normal array
    heapq.heapify(arr)
    n = len(arr)
    new_list = [0] * n

    for i in range(n):
        minn = heapq. heappop(arr)
        new_list[i] = minn
    return new_list


heapsort([1,3,5,7,9,2,4,6,8,8,0])
````

 # Maximum Sliding Window Problem and Solution Explanation

## Problem Explanation

The **Maximum Sliding Window** problem involves finding the maximum element in each contiguous subarray (window) of size `k` in a given array of integers `nums`. The window slides from the leftmost position to the rightmost position of the array, moving one element at a time. For each position of the window, we need to output the maximum element within that window.

### Problem Details
- **Input**:
  - `nums`: An array of integers (e.g., `[1, 3, -1, -3, 5, 3, 6, 7]`).
  - `k`: The size of the sliding window (e.g., `3`).
- **Output**:
  - An array containing the maximum element of each window of size `k` as the window slides from left to right.
- **Constraints**:
  - The window moves one position to the right each time.
  - You can only access the `k` elements in the current window.
  - The output array will have `n - k + 1` elements, where `n` is the length of `nums`.






## Solution Explanation

The provided solution uses a **deque** (double-ended queue) to efficiently track the indices of potential maximum elements in the sliding window. The deque maintains indices of elements in a way that ensures the front always points to the index of the maximum element in the current window, and elements in the deque are in **decreasing order** of their values. This approach achieves a time complexity of O(n) by avoiding redundant comparisons.



````python
__import__("atexit").register(lambda: open("display_runtime.txt", "w").write("0"))

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        n = len(nums)
        dq = deque()
        output = []
        start = 0
        for right in range(n):
            while dq and nums[right] >= nums[dq[-1]]:
                dq.pop()
            dq.append(right)

            if dq[0] < start:
                dq.popleft()

            if right + 1 >= k:
                output.append(nums[dq[0]])
                start += 1
            
        return output
````


### How the Solution Works

1. **Initialization**:
   - A deque (`dq`) is used to store indices of elements in the array.
   - An `output` list stores the maximum values for each window.
   - `start` tracks the left boundary of the sliding window.
   - `right` iterates through the array to represent the right boundary of the window.

2. **Main Logic**:
   - For each element at index `right`:
     - **Remove smaller elements**: While the deque is not empty and the current element (`nums[right]`) is greater than or equal to the element at the index at the back of the deque (`nums[dq[-1]]`), pop the back element. This ensures the deque maintains indices of elements in decreasing order, as smaller elements cannot be the maximum in the current or future windows.
     - **Add current index**: Append the current index `right` to the deque.
     - **Remove out-of-window indices**: If the index at the front of the deque (`dq[0]`) is outside the current window (i.e., `dq[0] < start`), pop it from the front.
     - **Record maximum**: Once the window size is at least `k` (i.e., `right + 1 >= k`), the element at the index at the front of the deque (`nums[dq[0]]`) is the maximum for the current window. Append it to the `output` list, and increment `start` to slide the window.

3. **Why Use a Deque?**
   - The deque stores indices of elements in decreasing order of their values, ensuring the front always holds the index of the maximum element in the current window.
   - By storing indices instead of values, we can easily check if an element is still within the window (using `start`) and access the actual value when needed.
   - The deque allows O(1) operations for adding/removing elements from both ends, making it ideal for this problem.

4. **Example Walkthrough** (for `nums = [1, 3, -1, -3, 5, 3, 6, 7], k = 3`):
   - Initialize: `dq = deque()`, `output = []`, `start = 0`.
   - **right = 0**: `nums[0] = 1`, `dq = [0]`.
   - **right = 1**: `nums[1] = 3 > nums[0] = 1`, pop `0`, `dq = [1]`.
   - **right = 2**: `nums[2] = -1 < nums[1] = 3`, `dq = [1, 2]`. Window size = 3, `output = [3]`, `start = 1`.
   - **right = 3**: `nums[3] = -3 < nums[2] = -1`, `dq = [1, 2, 3]`. `dq[0] = 1 >= start = 1`, `output = [3, 3]`, `start = 2`.
   - **right = 4**: `nums[4] = 5 > nums[1] = 3`, pop all, `dq = [4]`. `output = [3, 3, 5]`, `start = 3`.
   - **right = 5**: `nums[5] = 3 < nums[4] = 5`, `dq = [4, 5]`. `output = [3, 3, 5, 5]`, `start = 4`.
   - **right = 6**: `nums[6] = 6 > nums[4] = 5`, pop all, `dq = [6]`. `output = [3, 3, 5, 5, 6]`, `start = 5`.
   - **right = 7**: `nums[7] = 7 > nums[6] = 6`, pop all, `dq = [7]`. `output = [3, 3, 5, 5, 6, 7]`, `start = 6`.
   - Final output: `[3, 3, 5, 5, 6, 7]`.

### Time and Space Complexity

- **Time Complexity**: O(n)
  - Each element is pushed to and popped from the deque at most once.
  - All operations (push, pop, accessing elements) are O(1).
  - Total SD: Thus, the overall time complexity is O(n), where `n` is the length of `nums`.

- **Space Complexity**: O(k)
  - The deque stores at most `k` indices.
  - The output array requires O(n - k + 1) space, but this is part of the required output.
