````python
# Python program to demonstrate stack implementatio # using collections.deque

from collections import deque

stack = deque()

# append() function to push element in the stack
stack.append('a')
stack.append('b')
stack.append('c')

print('Initial stack:')
print(stack)

# pop() function to pop # element from stack in LIFO order
print('\nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack after elements are popped:')
print(stack)

# uncommenting print(stack.pop() # will cause an IndexError as the stack is now empty
````

### Implementation using queue module
-Queue module also has a LIFO Queue, which is basically a Stack. Data is inserted into Queue using the put() function and get() takes data out from the Queue. 

#### There are various functions available in this module: 

- maxsize - Number of items allowed in the queue.
- empty() - Return True if the queue is empty, False otherwise.
- full() - Return True if there are maxsize items in the queue. If the queue was initialized with - - maxsize=0 (the default), then full() never returns True.
- get() - Remove and return an item from the queue. If the queue is empty, wait until an item is available.
- get_nowait() - Return an item if one is immediately available, else raise QueueEmpty.
- get(item) - Put an item into the queue. If the queue is full, wait until a free slot is available before adding the item.
- get_nowait(item) - Put an item into the queue without blocking. If no free slot is immediately available, raise QueueFull.
- size() - Return the number of items in the queue.

````python
# Python program to
# demonstrate stack implementation
# using queue module

from queue import LifoQueue

# Initializing a stack
stack = LifoQueue(maxsize=3)

# qsize() show the number of elements
# in the stack
print(stack.qsize())

# put() function to push
# element in the stack
stack.put('a')
stack.put('b')
stack.put('c')

print("Full: ", stack.full())
print("Size: ", stack.qsize())

# get() function to pop
# element from stack in
# LIFO order
print('\nElements popped from the stack')
print(stack.get())
print(stack.get())
print(stack.get())

print("\nEmpty: ", stack.empty())
````



## 🔁 STACKS – LIFO (Last In, First Out)

### ✅ When to Use

Use a **stack** when you need to **remember a sequence in reverse**, **backtrack**, or **postpone decisions**.

### 🛠️ Common Operations

```python
stack = []
stack.append(10)   # push
stack.pop()        # pop
stack[-1]          # peek (top element)
```

### 🔎 Real-world & DS Use Cases

| Use Case                                                  | Why Stack is Best                               |
| --------------------------------------------------------- | ----------------------------------------------- |
| **Undo functionality** (in text editors, browsers)        | LIFO — the last action undone first             |
| **Function calls / Recursion**                            | Call stack keeps track of function calls        |
| **Balancing symbols** (e.g., `{[()]}`)                    | Stack matches open-close brackets in LIFO order |
| **DFS (Depth First Search)**                              | Stack handles traversal logic efficiently       |
| **Expression Evaluation (Postfix, Infix)**                | Stack processes operands/operators step-by-step |
| **Backtracking algorithms** (like maze solvers, N-Queens) | Stack allows rollback to previous decisions     |

### 🆚 Why Not Queue or List?

* Unlike queues, **stacks allow immediate access to the most recent item**, essential in nested calls or reversible actions.
* Plain lists can simulate stacks, but their semantic clarity and performance (O(1) for push/pop from end) make stacks preferable.

---

## 🪜 QUEUES – FIFO (First In, First Out)

### ✅ When to Use

Use a **queue** when **order of processing matters**, like **tasks arriving over time**.

### 🛠️ Common Operations

```python
from collections import deque
queue = deque()
queue.append(10)     # enqueue
queue.popleft()      # dequeue
```

### 🔎 Real-world & DS Use Cases

| Use Case                                    | Why Queue is Best                                |
| ------------------------------------------- | ------------------------------------------------ |
| **Task scheduling / Job queues**            | Processes tasks in the order they arrive         |
| **BFS (Breadth First Search)**              | Explores neighbors level-by-level                |
| **Producer-Consumer problems**              | Ensures items are processed in the correct order |
| **Network request handling / IO buffering** | Incoming packets or data streams use FIFO order  |
| **Printing tasks**                          | First document sent should be printed first      |
| **Rate limiting / Sliding Window**          | Maintains a time-based window efficiently        |

### 🆚 Why Not Stack or List?

* Stacks reverse order (bad for tasks), and lists (especially pop(0)) are **O(n)** — slower than `deque` which offers **O(1)** time.

---

## 🧩 How They Fit into Other Data Structures

| Data Structure                       | Use of Stack or Queue                                  |
| ------------------------------------ | ------------------------------------------------------ |
| **Trees**                            | BFS = Queue, DFS = Stack (recursive or iterative)      |
| **Graphs**                           | BFS = Queue, DFS = Stack                               |
| **Parsing expressions / compilers**  | Stacks help evaluate expressions, syntax tree building |
| **Dynamic Programming (Tabulation)** | Sometimes queues help in bottom-up traversal           |
| **Linked Lists**                     | Stacks for reverse traversal, Queues for simulation    |
| **Caches (LRU)**                     | Queues (usually Doubly Linked List + HashMap)          |

---

## ⚔️ Stack vs Queue – Use Case Showdown

| Scenario                              | Better Choice | Why                                         |
| ------------------------------------- | ------------- | ------------------------------------------- |
| **Web browser history**               | Stack         | You need "back" and "forward" functionality |
| **Customer service system**           | Queue         | First person in line should be served first |
| **Maze solving (backtracking)**       | Stack         | Go deep, and backtrack when needed          |
| **Shortest path in unweighted graph** | Queue         | BFS gives optimal shortest path             |
| **Function call resolution**          | Stack         | Nested calls return in reverse order        |
| **Operating system scheduler**        | Queue         | Round-robin scheduling with FIFO            |

Quesues has following operations basically


1. void push(int x) Pushes element x to the back of the queue.
2. int pop() Removes the element from the front of the queue and returns it.
3. int peek() Returns the element at the front of the queue.
4. boolean empty() Returns true if the queue is empty, false otherwise.


### Implement a first in first out (FIFO) queue using only two stacks. 
The implemented queue should support all the functions of a normal queue (push, peek, pop, and empty).

Implement the MyQueue class:

Queues operation => stack operation

push(int x)   => we use appen method of stack

pop() => first we peek the stack it gives us the first element  and then we pop out the element from the 2nd stack

 peek()=> check 2nd stack is empty or not and then pop from the 1st stack and append it to the 2nd stack  and then return the last elelment of the 2nd stack 

 empty() => check both stack is empty or not 
'''

````python
class MyQueue(object):
    def __init__(self):
        self.in_stk = []
        self.out_stk = []
	# Push element x to the back of queue...
    def push(self, x):
        self.in_stk.append(x)
	# Remove the element from the front of the queue and returns it...
    def pop(self):
        self.peek()
        return self.out_stk.pop()
	# Get the front element...
    def peek(self):
        if not self.out_stk:
            while self.in_stk:
                self.out_stk.append(self.in_stk.pop())
        return self.out_stk[-1]
	# Return whether the queue is empty...
    def empty(self):
        return not self.in_stk and not self.out_stk
````

# implement stack using queue

Implement the MyStack class:

1. void push(int x) Pushes element x to the top of the stack.

2. int pop() Removes the element on the top of the stack and returns it.
3. int top() Returns the element on the top of the stack.
4. boolean empty() Returns true if the stack is empty, false otherwise.

Just use a queue where you "push to front" by pushing to back and then rotating the queue until the new element is at the front (i.e., size-1 times move the front element to the back).


1. void push(int x) => push to the front by appendleft()
2. int pop() => pop from the left side 
3. int top() => return the 0th element by indexing 
4. boolean empty() => check the length 0of the queue


````python
from collections import deque
class Stack:

    def __init__(self):
        self._queue = deque()

    def push(self, x):
        q = self._queue
        q.append(x)
        for _ in range(len(q) - 1):
            q.append(q.popleft())
        
    def pop(self):
        return self._queue.popleft()

    def top(self):
        return self._queue[0]
    
    def empty(self):
        return not len(self._queue)
````
