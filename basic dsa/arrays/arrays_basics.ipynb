### methods of array 


```python
#Append any value to the array using append() method
my_array.append(6)  # value 6 was appended to the existing array values.

arr.append(9)



#  Insert value in an array using insert() method

my_array.insert(0,0) #first argument is the index while second argument is the value.
arr.insert(0,0)

# Extend python array using extend() method
my_extnd_array = array('i', [7,4,4,8,9,10])
my_array.extend(my_extnd_array)

arr.extend(my_extnd_array)

print(arr)
print(my_array)
```

```python
# Add items from list into array using fromlist() method
c=[19,20,21]
my_array.fromlist(c)
 ## arr = [] has no attribute fromlist so we use extend 
arr.extend(c)

print(arr)
print(my_array)
```

```python
#  Remove any array element using remove() method
my_array.remove(19)
arr.remove(4)

print(arr)
print(my_array)
```

```python
#  Remove last array element using pop() method
my_array.pop()
arr.pop()
my_array.pop(1) # we can give index here 

# Fetch any element through its index using index() method
print(my_array.index(0))
print(arr.index(7))

print(arr)
print(my_array)
```

```python
# Reverse a python array using reverse() method
my_array.reverse()
arr.reverse()

print(arr)
print(my_array)
#  Get array total element and buer information  through buer_info() method
```

# Slicing of an Array or list
- Elements from beginning to a range use [:Index]
- Elements from end use [:-Index]
- Elements from specific Index till the end use [Index:]
- Elements within a range, use [Start Index:End Index]
- Print complete List, use [:].
- For Reverse list, use [::-1].

- sort() - sort the complete list
- max() - gives maximum of the list
- min() - gives minimum of the list
- sum() - gives sum of all the elements in list for integer list
- index() - gives first index of the element specified
- append() - adds the element to the end of the list
- remove() - removes the element specified
