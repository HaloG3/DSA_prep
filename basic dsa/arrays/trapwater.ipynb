# trapwater classic hard question



Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.

```python
# solution 1 with help of numpy array 
# take an array height and find left maximum and right maximum 
# using the height given in questoin and 
# then take the minimum of both left and right maximum and convert it into integer and return it
# np.accumulate.maximum :  NumPy function that computes the cumulative maximum of elements in an
#  array along a specified axis. It returns an array where each element at index i is the maximum 
# value encountered from the start of the array up to and including index i.
#  This function is particularly useful for finding the running maximum over a sequence of values
import numpy as np
class Solution:
    def trap(self, height):
        height = np.array(height)
        leftmax = np.maximum.accumulate(height)
        rightmax = np.maximum.accumulate(height[::-1])[::-1]
        return int(np.sum(np.minimum(leftmax, rightmax) - height))
```

```python
class Solution:
    def trap(self, height: List[int]) -> int:
        left=0
        right=len(height)-1
        leftmax=0
        rightmax=0
        total=0
        while left<right:
            if height[left]<height[right]:
                if height[left]>=leftmax:
                    leftmax=height[left]
                else:
                    total+=(leftmax-height[left])
                left+=1
            else:
                if height[right]>=rightmax:
                    rightmax=height[right]
                else:
                    total+=(rightmax-height[right])
                right-=1
        return total
```
