# Diameter of a tree 


```python
# class solution
def diameterofBinaryTree(self, root):

    res = 0 # we need to define member function of class / global function  
    def dfs(curr):
        if not curr:# check the height 
            return 0
        left = dfs(curr.left)
        right = dfs(curr.right)
        nonlocal res  # tell it that it is non local 
        res = max(res,left+right)
        return 1+max(left,right) # dont return the height 
    dfs(root)
    return res 

# t: = O(n) = space  for non balanced tree 
# t:= O(n) s: O(n log n) for balanced tree
```

# print the nth level of the tree 
- Tree in the image below, K = 2 [9,6]


```python
# Python code for the above approach
class Node:

    def __init__(self, d):
        self.data = d
        self.left = None
        self.right = None

# Recursive function to print all
# nodes of a Binary Tree at a
# given level using DFS traversal
def printNodes(root, level, K):

    # Base Case
    if (root == None):
        return

    # Recursive Call for
    # the left subtree
    printNodes(root.left, level + 1, K)

    
    # the right subtree
    printNodes(root.right, level + 1, K)

   
    # the required level
    if (K == level):
        print(root.data, end=" ")

# Driver Code
root = Node(3)
root.left = Node(9)
root.right = Node(6)
root.left.left = Node(11)
K = 2

printNodes(root, 1, K)
```
