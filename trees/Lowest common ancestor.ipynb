
## 236 The Lowest Common Ancestor (LCA) 
of two nodes in a binary tree is the deepest node that has both of those nodes as descendants. In simpler terms, it's the shared ancestor of two nodes that is furthest from the root of the tree

we use DFS here; 
- check the current node is equal to p,q
- specific condition : there must exist p and q in the binary tree
- if we found any p or q then return to back and explore other nodes as we do in DFS 
- and if we found any p or q in other node then return back and this node is the lowest common ancestor
- if we not found any non - None node then we return back and do dfs means explore other node
## time = O(N)  
- in worst case we need to explore the whole tree so it is equal to the node of the tree
## space = O(1)
- we not creating any extra space so it is 1 if we are not considering the recursive stack framing otherwise it is also **O(N)**

```python
#class solution
def LCA(self, root, p , q):
    if not root:
        return None
    if root == p or root == q :
        return root
    l = self.LCA(root.left, p , q)
    r = self.LCA(root.right, p, q)

    if l and r : 
        return root 
    else:
        return l or r
```
