```python
# by mermoisation method 
# Fibonacci using Memoization (Top-Down Dynamic Programming)
def fibonacci_memo(n, memo={}):
    # Base cases
    if n == 0:
        return 0
    if n == 1:
        return 1

    # If result already computed, return from memo
    if n in memo:
        return memo[n]

    # Otherwise compute and store in memo
    memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
    return memo[n]

# Print the Fibonacci series up to n terms
def print_fibonacci_memo(n):
    print("Fibonacci Series using Memoization:")
    for i in range(n):
        print(fibonacci_memo(i), end=" ")
    print()

# Example usage
print_fibonacci_memo(10)
```

```python
# Fibonacci using Tabulation (Bottom-Up Dynamic Programming)
def fibonacci_tab(n):
    if n == 0:
        return []

    # Initialize a list to store Fibonacci numbers up to n
    fib = [0] * n
    fib[0] = 0
    if n > 1:
        fib[1] = 1

    # Build the Fibonacci series from bottom up
    for i in range(2, n):
        fib[i] = fib[i - 1] + fib[i - 2]

    return fib

# Print the Fibonacci series up to n terms
def print_fibonacci_tab(n):
    print("Fibonacci Series using Tabulation:")
    series = fibonacci_tab(n)
    for num in series:
        print(num, end=" ")
    print()

# Example usage
print_fibonacci_tab(1)
```
