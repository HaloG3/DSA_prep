## solution approach
- First, we can calculate a **prefix sum** array s for the given nums array. The prefix sum array contains the sum of all numbers up to and including the ith index of nums. This helps us calculate the sum of any subarray in constant time.

- we use 2 loops for each j:
1. inner loop over i, for i in range(1 ,j-1), 
- We check if the sum of elements from 0 to i-1 is equal to the sum from i+1 to j-1
- and we use dict() to make the mark the set seen if we found 
2.  outer loop over k, for k in range(j+2,n-1)  
- We check if the sum of elements from j+1 to k-1 is equal to the sum from k+1 to n-1.
- If it is and the sum is already in seen, there exists at least one combination of i, j, k that satisfies all conditions, thus, we return true

### prepare prefix sums and hashmap for this question



```python
 #solution form web : https://algo.monster/liteproblems/548
from typing import List

class Solution:
    def splitArray(self, nums: List[int]) -> bool:
        # Get the length of the input array
        n = len(nums)
      
        # Initialize a prefix sum array with an extra position for simplicity
        prefix_sum = [0] * (n + 1)
      
        # Compute the prefix sum array where prefix_sum[i] represents
        # the sum of elements from nums[0] to nums[i-1]
        for i, num in enumerate(nums):
            prefix_sum[i + 1] = prefix_sum[i] + num
      
        # The main loop to check for split positions starting at index 3
        # and ending at n-4 to ensure there are enough elements on both sides
        for mid in range(3, n - 3):
            # Store sums that can be created from the first half of the array
            seen_sums = set()
          
            # Check for all possible splits in the first half
            for left in range(1, mid - 1):
                # If a valid split is found, add the sum to the set
                if prefix_sum[left] == prefix_sum[mid] - prefix_sum[left + 1]:
                    seen_sums.add(prefix_sum[left])
          
            # Check for all possible splits in the second half of the array
            for right in range(mid + 2, n - 1):
                # Check if there is a valid split that matches any sum in 'seen_sums'
                if (prefix_sum[n] - prefix_sum[right + 1] == prefix_sum[right] - prefix_sum[mid + 1] 
                        and prefix_sum[n] - prefix_sum[right + 1] in seen_sums):
                    return True
      
        # If we reach this point, no valid split was found
        return False
    
```

####  the time complexity
-  of the code is O(n^2) due to the nested loops.
##### the overall space complexity 
- of the function is O(n).
